```python
temperature_data["Date"]=pd.to_datetime(temperature_data["Date"])
```

```python
temperature_df_sort_by_date=temperature_data.sort_values(by="Date")
```

```python
temperature_pivot_df=temperature_df_sort_by_date.pivot_table(values="Data_Value",index="Date",columns="Element")
```

```python
temperature_pivot_df
```

```python
temperature_pivot_df["Day"]=temperature_pivot_df.index.day
temperature_pivot_df["Month"]=temperature_pivot_df.index.month
temperature_pivot_df["Year"]=temperature_pivot_df.index.year
```

```python
temperature_pivot_df
```

```python
temperature_pivot_df[(temperature_pivot_df["Month"]==2) & (temperature_pivot_df["Day"]==29)]
```

```python
temperature_df_without_leapyear=temperature_pivot_df[(temperature_pivot_df["Month"]!=2) & (temperature_pivot_df["Day"]!=29)]
```

```python
temperature_df_without_leapyear
```

```python
temperature_df_without_leapyear[(temperature_df_without_leapyear["Month"]==2) & (temperature_df_without_leapyear["Day"]==29)]
```

```python
temperature_df_without_leapyear.dtypes
```

```python
temperature_df_from_05_to_14=temperature_df_without_leapyear[temperature_df_without_leapyear["Year"]!=2015]
```

```python
temperature_df_from_05_to_14
```

```python
Minimum_temperature_df_05_to_14=temperature_df_from_05_to_14[["Day","TMIN"]]
```

```python
Minimum_temperature_df_05_to_14
```

```python
Maximum_temperature_df_05_to_14=temperature_df_from_05_to_14[["Day","TMAX"]]
```

```python
Maximum_temperature_df_05_to_14
```

```python
temperature_df_of_15=temperature_df_without_leapyear[temperature_df_without_leapyear["Year"]==2015]
```

```python
temperature_df_of_15
```

```python
# graph drawing 
```

```python
%matplotlib inline 
plt.figure(figsize=(20,12))

plt.plot(Maximum_temperature_df_05_to_14.values,c="blue",label="High temperature record")
plt.plot(Minimum_temperature_df_05_to_14.values,c="red",label="Low temperature record")

plt.xlabel("day")
plt.ylabel("temperature")
plt.title("2014-15 temperature record")

plt.show()
```
